```python
fn = "timing_test_1_results.csv"
```

```python
data = np.loadtxt(fn, delimiter=",")
```

```python
def two_cols_to_int(col1, col2):
    myvect = data[:,col1]*256+data[:,col2]
    return myvect
```

```python
enc = two_cols_to_int(0,1)
```

```python
plt.figure()
plt.plot(enc)
```
